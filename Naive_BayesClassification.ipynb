```python
class naivebayes:
  def __init__(self,dataset,target):
    self.dataset=dataset
    self.target=target
    self.values,self.counts = np.unique(dataset[target], return_counts=True)
  def probability(self,feature,targetvalue,val):
    #values=dataset[target].unique()
    feature_values=self.dataset[feature].unique()
    y={i:0 for i in feature_values }
    k=0
    for i in self.dataset[feature].values:
      #print("feature][k]==targetvalue   =  ",feature][k]==targetvalue)
      if self.dataset[feature][k]==targetvalue and self.dataset[target][k]==val :
        y[i]=y[i]+1
      k=k+1
    return y[targetvalue]


  def classify(self,example):
      n=len(example)
      h=[]
      for i in range(len(self.values)):
        p=self.counts[i]
        for j in range(n):
          k=self.probability(self.dataset.columns[j],example[j],self.values[i])
          #print("k=",k)
          p=p*(k/self.counts[i])
        #print("probability of ",self.values[i],"is:" ,p) 
        h.append(p)
      total=sum(h)
      for i in range(len(h)):
        h[i]=h[i]/total
        print("probability of ",self.values[i],"is:" ,h[i])
      #print(h) 
      print("The Class Label is:",self.values[np.argmax(h)])
```

```python
test = ['rain','cool','normal','strong']
ob=naivebayes(df,target)
k=ob.classify(test)
k
```
